```python
import requests #collects the webpage data object 
from bs4 import BeautifulSoup #converts the requests object in readable object
```

```python
def load_soup(url):
    page = requests.get(url) #collects data from webpage
    if page.status_code == 200:
        return BeautifulSoup(page.text)
```

```python
url= 'https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_2_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_2_0_na_na_na&as-pos=2&as-type=TRENDING&suggestionId=mobiles&requestId=c43b8db4-e795-48b4-9c96-e97a842bd204' #webpage address
soup = load_soup(url) #collects data from webpage
if soup is not None:
    print('Data loaded')
else:   
    print('Data could not be loaded') 
```

```python
target = soup.find('div',attrs={'class':'_1YokD2 _3Mn1Gg'})
if target is not None:
    print('Target area is found')
else:
    print('Targeted area could not be found')    
```

```python
items = target.find_all('div', class_ = '_2kHMtA')
if items is not None:
    print(len(items), 'items found in page')
else:    
    print(len(items), 'items does not found in page')
```
